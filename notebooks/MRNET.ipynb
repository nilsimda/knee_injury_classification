```python
#imports
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import torch
```

```python
DATA_PATH = os.path.join("..", "data", "MRNet-v1.0")

def show_image(image, dataset=DATA_PATH):
    image_path = os.path.join(dataset, image)
    img_arr = np.load(image_path)
    plt.imshow(img_arr[0])
    plt.axis('off')
```

```python
axial_files = sorted(os.listdir(os.path.join(DATA_PATH, "train", "axial")))
coronal_files = sorted(os.listdir(os.path.join(DATA_PATH, "train", "coronal")))
sagittal_files = sorted(os.listdir(os.path.join(DATA_PATH, "train", "sagittal")))

plt.figure(figsize=(16,16))
plt.subplot(131)
show_image(os.path.join("train", "axial", axial_files[1]))
plt.title("Axial")

plt.subplot(132)
show_image(os.path.join("train", "coronal", coronal_files[1]))
plt.title("Coronal")

plt.subplot(133)
show_image(os.path.join("train", "sagittal", sagittal_files[1]))
plt.title("Sagittal");
```
